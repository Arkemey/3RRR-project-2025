```python
import time
from pydxl import Mx28, SerialLink

link = SerialLink(device="/dev/ttyUSB0", baudrate=1_000_000, protocol_version=1.0)
servo = Mx28(identifier=1, serial_link=link)

# Ping / check
servo.ping()

# Set LED (optional)
servo.led = True

# Enable torque
servo.torque_enable = True

# Move
servo.goal_position = 2000
time.sleep(2)

servo.goal_position = 1500
time.sleep(2)

# Disable
servo.torque_enable = False

link.close()
```
